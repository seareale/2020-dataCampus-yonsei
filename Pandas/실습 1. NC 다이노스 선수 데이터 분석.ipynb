```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
```

# NC 다이노스 선수 데이터 분석
## 1. 데이터 적재 및 통합

```python
NCAll = pd.read_excel('data/NC Dinos.xlsx', sheet_name=None)
NC13, NC14, NC15 = NCAll.values()
```

```python
# NC13, NC14, NC15에 각각 년도 컬럼 추가
NC13['년도'] = 2013
NC14['년도'] = 2014
NC15['년도'] = 2015
```

```python
NC13.head()
```

```python
# Unnamed 삭제
NC13.drop('Unnamed: 0', axis = 1, inplace = True)
NC14.drop('Unnamed: 0', axis = 1, inplace = True)
NC15.drop('Unnamed: 0', axis = 1, inplace = True)
```

```python
NC13.head()
```

```python
# NC13, NC14, NC15 하나로 합치기
NC = pd.concat([NC13, NC14, NC15], axis = 0, ignore_index = True)
```

## 2. 2015년도 선수 기록을 '타율'로 내림차순 정렬

```python
NC15.sort_values(by='타율', ascending = False).head()
```

```python
# 1. 타율이 같으면 타석수가 많은 선수부터 출력
NC15.sort_values(by = ['타율', '타석'], ascending = [False, True]).head()
```

```python
NC15[NC15.타석 >= 200].sort_values(by = '타율', ascending = False).head()
```

## (실습 #1) 2013~2015년  선수별로 3년 간 평균 안타수를 구한 후, 평균 안타수가 가장 많은 상위 10명의 선수를 출력

```python
# NaN 값이 있는지 확인
NC.info()
```

```python
#NC.groupby('선수명')['안타'].mean()

ex1 = NC.pivot_table(index = '선수명', aggfunc='mean', values='안타').sort_values(by = '안타', ascending = False).head(10)
ex1
```

## (실습 #2) 2013~2015년 3년 동안 선수들의 평균 안타수, 총 안타수, 선수로 뛴 시즌수 구하기
선수로 뛴 시즌수 = 해당 시즌의 선수 목록에 없다면 그 선수는 해당 시즌에 선수에 포함되지 않은 것임. 

```python
# aggfunc dictionary 형식으로 value에 대한 함수 지정 가능
data = NC.pivot_table(index = '선수명', aggfunc ={'안타' : ['mean', 'count', 'sum']}, values = '안타')

#data = NC.pivot_table(index = '선수명', aggfunc =['mean', 'count', 'sum'], values = '안타')
data.columns = ['평균안타수', '시즌수', '총안타수']
data.sort_values(by ='평균안타수', ascending = False).head()
```

## (실습 #3) 실습 #1에서 구한 상위 10명의 타자들의 3년간 안타, 홈런, 타율을 출력하기

```python
# 1. 필요한 컬럼만 선택
ex3 = NC[['선수명', '안타', '홈런','타율','년도']]
ex3.head()
```

```python
# 2. 상위 10명 선택, 인덱스 지정, Unstack
ex3[ex3.선수명.isin(ex1.index)].set_index(['선수명','년도']).unstack(1).fillna('-')


```
