```python
# 인자에 * 붙이면 입력받은 많은 인자를 튜플로 받음
def min_max(*x):
    
    max_ = max(x)
    min_ = min(x)
    
    answer = (min_, max_)
    
    return answer

min_max(9, -1, 100, 3.2,  -5.2, 0)
```

```python
# 인자에 * 붙이면 입력받은 많은 인자를 딕셔너리로 받음
def vitamin_c(**fruits):

    total_C = sum(fruits.values())
    
    print('-'*40)
    print('Fruit Name',' '*11, 'Vitamin C Content')
    print('-'*40)
    
    for i in fruits:
       # print(i, ' '*(len('Fruit Name')+len(' '*11)-len(i)), fruits[i]) 별루
        print('{:23}{:2}'.format(i, fruits[i]))
        
    print('-'*40)
    
    return total_C

print('총 함유량:', vitamin_c(citron=105,strawberry=94,quince=81,lemon=70,tangerine=39))
print('\n\n')
print('총 함유량:', vitamin_c(orange=43, kiwi=27, cherryTomato=21, apple=4))
# 함수 실행 순서를 생각해봐. vitamin_C()가 '총 함유량'보다 먼저 실행되겠지?
# vitamin_C 안의 내용들이 수행된 후, print('총 함유량', total_C) 가 수행됨,
```

```python
# 1
def ch_bin(x):
    d = {}
    # count를 쓰면 문자열 길이만큼 카운팅을 실시. 데이터 처리량이 많이지면 매우매우 비효울적임
    for i in x:
        if i not in d:
            d[i] = 1
        else :
            d[i] += 1
    return d


def print_dic(a):
    for x in a:
        print("{} {}".format(x,a[x]))

def find_ch(d,n):
    for k in d:
        if d[k] == n:
            return k
        
def inv_dic(d):
    new = {}
    for x in d:
        if d[x] not in new: # 기존 key 값을 list형태로 넣기
            new[d[x]] = [x]
        else :
            new[d[x]].append(x)
    return (new)
    

num = eval(input("문자열 : "))
d = ch_bin(num)
print_dic(d)
print(find_ch(d,2))
print_dic(inv_dic(d))
```

```python
# 3

def my_dist(x1, y1, x2, y2):
    # math 쓸지 안쓸지 모른다.
    import math  # math안의 모든 내용을 로드해서 메인메모리에 올린다.
    # from math import sqrt  -> sqrt만 불러옴, sqrt그대로 사용 가능
    return math.sqrt((x1-x2)**2 + (y1-y2)**2)

print(my_dist(1, 0, 2, 0))
```
