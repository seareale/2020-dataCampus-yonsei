# 1. 개요

다변량 자료  
- 키와 몸무게의 관계와 같이 두개 이상의 변수를 동시에 다루어야 하는 자료
- 두개인 경우를 특히 이변량 자료라고 한다
- 일변량 자료는 vector 에 저장하여 분석할 수 있고, 다변량 자료는 matrix 또는 data frame 에 저장하여 분석한다

변수(variable)
- ex 키, 몸무게
- 변수는 데이터셋에서 열로 표현된다 (Sepal.Length, Sepal.Width, Petal.Length, Petal.Width)

관측값(observation), 레이블(label)


# 2. 산점도(Scatter plot)

이변량 자료의 분포 및 상관관계를 시각적으로 확인
 - mtcars 데이터셋에서 자동차 중량(wt) 와 연비(mpg) 의 상관관계를 산점도로 확인

```r
wt <-mtcars$wt
mpg <- mtcars$mpg
plot(wt, mpg,                  # 2개 변수(x축,y축)     
     main="Car Weight-mpg",    # 제목
     xlab="Car Weight ",       # x축 레이블
     ylab="Miles Per Gallon ", # y축 레이블
     col="red",                # point 의 color
     pch=19)                   # point 의 종류  
```

중량이 큰 차일수록 연비가 떨어지는 것을 관찰할 수 있다

## 2.3 pairs()

pairs() : 여러 변수들 사이의 상관관계를 한번에 확인

# 3. 상관 분석

두 변수 X와 Y 간의 선형성의 정도를 측정하는 통계량으로 다음과 같이 정의 됨



일반적으로
- -1 <= r <= 1
- r > 0 : 양의 상관 관계
- r < 0 : 음의 상관관계
- 1 이나 -1 에 가까울수록 상관성이 높다

## 3.1 예제

음주 정도와 혈중 알코올 농도의 상관도 분석 


```r
beers = c(5,2,9,8,3,7,3,5,3,5)
bal = c(0.1,0.03,0.19,0.12,0.04,0.0095,0.07,
        0.06,0.02,0.05)
tbl = data.frame(cbind(beers,bal))  # 두 벡터를 컬럼(열) 방향으로 합친다  
                                    # rbind() : 두 벡터를 행 방향으로 합친다
tbl; class(tbl) 
plot(bal~beers,data=tbl)    # 산점도
                            # plot(tbl), plot(tbl[,1],tbl[,2]) 도 같은 결과 출력

res=lm(bal~beers,data=tbl) # 회귀식 도출
                            # 산점도를 가장 잘 표현할 수 있는 선형 모델(회귀식)을 구한다.

abline(res)                 # 회귀선그리기

cor(beers,bal)             # 상관성 분석 시행
                           # 두 벡터자료로 부터 상관계수를 계산한다.
```

- 여러 변수들간의 상관 계수를 동시에 구하는 방법

```r
cor(iris[,1:4])    # 4개 변수간 상관성 분석
```

```r
cor(iris[,1],iris[,3])
```

1. 다음은 10명의 수입과 교육받은 기간을 조사한 표이다. 수입과 교육기간 사이에 어느정도 상관관계가 있는지 조사하시오 (산점도, 상관계수 구하기)


```r
Income = c(125000,100000,40000,35000,41000,29000,35000,24000,50000,60000)
Years = c(19,20,16,16,18,12,14,12,16,17)
tbl = data.frame(cbind(Income,Years))

plot(Income~Years, data=tbl)
res = lm(Income~Years, data=tbl)
abline(res) # x, y 축 서로 다르면 안그려짐
cor(Income,Years)
```

# 4. 선그래프

연도별 인구증감 추이와 같이 시간 순서에 따른 데이터의 시각화에 많이 사용됨

## 4.1 선 하나인 그래프

월별 지각생 통계


```r
month = 1:12
late  = c(5,8,7,9,4,6,12,13,8,6,6,4)
```

월별 지각생 통계의 선 그래프

```r
plot(month,             # x data
     late,              # y data
     main="Late students",
     type= "l",         # 그래프의 종류 선택(알파벳)
                        # b : 점과선,      s : 계단형,    S : 계단형2
                        # o : 점위의 선,   h : 값에 해당하는 수직선
     
     lty=1,             # 선의 종류(line type) 선택
     lwd=1,             # 선의 굵기 선택
     xlab="Month ",     # x축 레이블
     ylab="Late cnt"    # y축 레이블
)
```

## 4.2 복수의 선그래프

1반, 2반의 지각생 데이터  



```r
month = 1:12
late1  = c(5,8,7,9,4,6,12,13,8,6,6,4)
late2  = c(4,6,5,8,7,8,10,11,6,5,7,3)
```

```r
plot(month,             # x data
     late1,              # y data
     main="Late students",
     type= "b",         # 그래프의 종류 선택(알파벳). 
     lty=1,                            # 선의 종류(line type) 선택
     col="red",          # 선의 색깔 선택          
     xlab="Month ",      # x축 레이블
     ylab="Late cnt"     # y축 레이블
)
lines(month,             
     late2, 
     type = "b", 
     col = "blue")
```

하나의 선그래프를 그린 후에 그 위에 또다른 선 그래프를 겹쳐 그리는 방식, 이외에 그리는 방법 많음
