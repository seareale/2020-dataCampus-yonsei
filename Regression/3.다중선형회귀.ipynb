# 1. 다중선형회귀

- 독립변수가 2개 이상이고 종속변수도 1개인 경우, 그들 간의 관계를 선형적으로 파악하는 회귀 방식
- 독립변수들 X1, X2, X3, ..., Xn과 종속변수 Y의 관계를 **Y = w0 + w1X1 + w2X2 + w3X3 + ... + wnXn** 형태의 1차 함수식으로 표현할 수 있다.

- 단순 선형 회귀에서 **독립변수의 개수만 늘어난 것(2개 이상)**이다.
- 따라서 단순 선형 회귀와 **동일한 절차**를 이용하여 분석을 수행할 수 있다.
- 단, 변수의 수가 많아지므로 이로 인해 발생할 수 있는 경우들을 고려해서 적절한 조치를 취해야 한다.

- 회귀 분석의 목적은 종속변수를 가장 잘 설명하는 독립변수들의 성향/특징을 찾아내어 이를 기반으로 기존의 자료를 설명하거나 새로운 결과를 예측하는 것이다.
- 즉, 독립변수 일부를 임의로 누락시키는 것은 해당 모형의 설명력이 낮아지는 문제를 불러올 수 있다.
- 따라서 통계(분석)적으로 회귀 분석을 수행하는 경우, 관련있는 **독립변수들을 일단 가급적 모두 고려**하는 것이 바람직하다.

## 1.1. 다중 공선성 (Multi-collinearity) 문제

- 독립변수들 간에 강한 상관 관계가 나타나는 것을 **다중 공선성**이라고 한다.   
(즉, 어떤 독립변수의 값이 독립적이지 않고 다른 독립변수(들)의 값에 의해서 결정된다는 것이다.)  
  
  
- 독립변수들끼리의 상관 계수 R이 높으면 (아무리 설명력이 좋다고 하더라도) 회귀 모형이 유의미하다고 보기 어렵다.
- 따라서 상관 계수가 높은 변수들을 삭제하거나, 주성분 분석(PCA) 기법 등을 이용하여 의존적인 성분을 제거한 뒤 회귀 분석을 수행해야 한다.

## 1.2. 상관 계수 R

- 공분산을 각각의 표준편차로 나누어 정규화한 수치
    - 변수 X, Y에 대하여 각각의 크기(단위)에 영향을 받지 않도록 단위를 보정한 것이라고 볼 수 있다.
        ![image.png](img/ch3_1.png)
    - **<span class="mark">상관 계수의 값은 -1 ≤ R ≤ 1이며, 1에 가까울수록 강한 양(+) 의 상관 관계, -1에 가까울수록 강한 음(–)의 상관관계이다. 0이면 서로 상관 관계가 없다</span>**

## 1.3. 독립변수들의 설명력과 수정된 결정 계수

factor가 많아지면 설명력이 높아진다. 독립변수가 많을수록 종속변수에 대한 설명력이 높아진다.

- 독립변수의 개수가 많아질수록 그 변수들이 종속변수에 끼치는 영향력은 늘어나게 된다. (<span class="mark">즉, 독립변수가 많을수록 종속변수에 대한 설명력은 증가한다.)</span>
- 따라서 다중 회귀 분석에서는 결정 계수 R2의 값이 단순회귀보다 높게 나오는 경향이 있고, 이는 독립변수의 수가 많아질수록 더욱 증가한다.
- 이를 보완하여 수정된 결정 계수를 도입했으며, 다중 회귀에서는 일반적으로 **수정 결정 계수의 값을 이용하여 분석 결과를 판단**한다.

### 1.3.1 결정 계수 R2

- 회귀식이 얼마나 설명력이 있는지 (즉, 얼마나 정확한지) 나타내는 지표이다.
![image.png](img/ch3_2.png)


- 결정 계수의 값은 0 ≤ R2 ≤ 1이며, 1에 가까울수록 설명력이 강하고 0에 가까울수록 설명력이 약하다.
- 일반적으로 결정 계수 R2의 값이 0.65 (65%) 이상이면 설명력이 있다고 판단한다.

### 1.3.2 수정 결정 계수 (Adjusted R2)

R2 보다 작은 값으로 만들도록 한다

- 결정 계수의 <span class="mark">값이 커지는 것을 보정</span>하기 위해 <span class="mark">데이터(표본)의 크기와 독립변수의 개수를 고려</span>하여 계산한 지표이다.
![image.png](img/ch3_3.png)


- 수정 결정 계수의 값은 항상 결정 계수 R2보다 작거나 같은 값이다.
- (모든 독립변수가 아니라) <span class="mark">종속변수에 영향을 주는 독립변수들만으로 설명되는 분산의 비율</span>이라고 볼 수 있다.

### 1. 데이터 선정 및 분포 정보 확인

- 보스턴 주택 가격 데이터를 불러와서 정보를 확인한다

```python
import sklearn.datasets as d

boston = d.load_boston()
print(boston.DESCR)
```

- 데이터를 DataFrame으로 변환한다.

```python
import pandas as pd

# 특성(독립변수)들을 DataFrame으로 변환
boston_df = pd.DataFrame(boston.data, columns = boston.feature_names)

# 종속변수를 덧붙인다.
boston_df["PRICE"] = boston.target

# 데이터의 형태와 크기를 확인한다.
print(boston_df.shape)

# 데이터의 일부를 확인해 본다.
boston_df.head()
```

- 컬럼 별로 종속변수와의 관계에 대한 산점도를 그린다.

```python
import matplotlib.pyplot as plt

# matplotlib.pyplot 가 먼저 import 되어 있어야한다.
import seaborn as sns 
```

```python
%matplotlib inline
# 3행 4열의 subplot을 준비한다.
fig, axs = plt.subplots(nrows=3, ncols=4, figsize=(15,10))
# fig -> subplot들이 들어간 box 전체 크기, 여기서 사용안함
# 안써도 할당해줘야함

# 전체 컬럼들에서 종속변수 및 불필요한 컬럼을 뺀다.
features = [f for f in list(boston_df.columns)[:-1] if f != "CHAS"]
features.sort()

# 0, AGE
# 1, B
# 2, CRIME
for i, feature in enumerate(features):
    # 현재 subplot의 행/열 번호를 갱신한다.
    r = int(i/4)
    c = i % 4
    # 산점도와 회귀선을 그린다.
    sns.regplot(x = feature, y="PRICE", data=boston_df, ax=axs[r][c])
    
plt.show()
```

### 2. 사이킷런으로 다중 선형 회귀 수행

- 데이터를 학습용과 검증용으로 분리한다.

```python
import sklearn.model_selection as ms

# DataFrame에서 독립변수와 종속변수를 다시 구분한다.
X = boston_df.drop(["PRICE"], 1)   # drop() 특정 column이나 row를 제외, 0:행, 1:열
y = boston_df["PRICE"]

# 학습용 및 검증용 데이터로 분리한다.
X_train, X_test, y_train, y_test = \
ms.train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
X_train.head()
```

- 최소제곱법으로 학습, 예측, 평가를 진행한다.

```python
import sklearn.linear_model as lm
import sklearn.metrics as mt
import numpy as np

# 훈련 데이터로 학습을 수행한다.
reg = lm.LinearRegression().fit(X_train, y_train)

# 검증 데이터로 예측을 수행한다.
y_pred = reg.predict(X_test)

# 평가 지표 값들을 계산한다.
mse = mt.mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = mt.r2_score(y_test, y_pred)

# 평가 지표 값들을 출력한다.
print("MSE : {:.3f}\nRMSE : {:.3f}\nR2 : {:.3f}".format(mse,rmse,r2))
```

- 추가로 수정 결정 계수를 계산한다

```python
# 데이터의 크기, 즉 레코드 수
n = len(X_train)

# 독립변수의 개수
k = len(X_train.columns)

# 수정된 결정 계수 값을 계산한다.  
adj_r2 = 1 - ((1-r2) * (n-1) / (n-k-1))    # 스탯츠모델에는 계산함수 있음

# 수정 결정 계수 값을 출력한다.
print("Adjusted R2 : {:.4f}".format(adj_r2))
```

    분포 정확도 70%를 의미

- 회귀 계수와 절편 값을 확인한다.

```python
# ndarray의 실수값들을 부동소수점으로 출력되도록 옵션을 변경한다.
np.set_printoptions(suppress = True)

# 회귀 계수들 적절히 반올림하여 출력한다.
print("회귀 계수 :", np.round(reg.coef_,3))

# 절편 값을 적절히 반올림하여 출력한다.
print("절편 :", round(reg.intercept_,3))

# ndarray의 실수값 출력 옵션을 원래대로 되돌린다.
np.set_printoptions(suppress=False)
```

- 회귀 계수를 각 독립변수 이름과 함께 정렬하여 확인한다.

```python
# 각 독립변수 이름이 인덱스가 되도록 회귀 계수를 Series로 생성한다.
coefs = pd.Series(data = np.round(reg.coef_,3), index=X.columns)

# 생성한 Series를 값의 내림차순으로 정렬한다.
coefs_sort = coefs.sort_values(ascending = False)

# 결과를 확인한다.
print(coefs_sort)
```

- 실제값과 예측값의 분포 차이를 플롯으로 표현해 본다.

```python
plt.figure(figsize=(6,6))

# s (마커 size)  -> 기본값 50 정도
plt.scatter(y_test, y_pred, marker="x", s=60)
```

### 3. 스탯츠모델로 다중 선형 회귀 수행

- 최소제곱법으로 학습, 예측, 평가를 진행한다.
    - 파이썬 sklearn에는 없는 R의 기능 사용, summary 등 유용  
    - 스탯츠모델에서의 R2, adj_R2는 test 데이터만이 아닌 전체 데이터로 구한것 
    - Prob (F-statistic) (F검정에 대한 p-value, 전체 모형에 대한 것)가 0.05 이상이면 각각의 회귀계수에 문제가 있는 것이 아닌 모형 자체에 문제가 있다, 설득력이 없다.
    - 각각의 회귀계수별로 P>|t| 존재, P>|t| 값이 0.05 이상이면 해당 회귀계수는, 독립변수는 무의미함. 

```python
import statsmodels.api as sm

# 훈련 데이터로 학습을 수행한다.
# .add_constant() 안하면 절편이 안나온다.
X_train = sm.add_constant(X_train)
reg = sm.OLS(y_train, X_train).fit()

# 검증 데이터로 예측을 수행한다.
X_test = sm.add_constant(X_test)
y_pred = reg.predict(X_test)

# 평가 지표 값들을 출력한다.
print(reg.summary())
```

        수정 결정 계수, 회귀 계수, 절편 값 등을 확인한다.

```python
# 위 데이터를 기반으로 INDUS, AGE 제외 

import sklearn.model_selection as ms

# DataFrame에서 독립변수와 종속변수를 다시 구분한다.
X = boston_df.drop(["INDUS", "AGE", "PRICE"], 1)   # drop() 특정 column이나 row를 제외, 0:행, 1:열
y = boston_df["PRICE"]

# 학습용 및 검증용 데이터로 분리한다.
X_train, X_test, y_train, y_test = \
ms.train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
import statsmodels.api as sm

# 훈련 데이터로 학습을 수행한다.
X_train = sm.add_constant(X_train)
reg = sm.OLS(y_train, X_train).fit()

# 검증 데이터로 예측을 수행한다.
X_test = sm.add_constant(X_test)
y_pred = reg.predict(X_test)

# 평가 지표 값들을 출력한다.
print(reg.summary())
```

# 2. 복습

```python
import sklearn.datasets as d

# dir(d)
# load_diabetes : 선형회귀분석 하기 좋음?

diab = d.load_diabetes()
```

```python
print(diab.DESCR)
```

```python
X = diab.data
y = diab.target
```

```python
len(X)
```

```python
import sklearn.model_selection as ms

# 학습용 데이터가 테스트 데이터보단 많아야한다.
X_train, X_test, y_train, y_test = \
ms.train_test_split(X,y, test_size=0.3, random_state = 2)
```

```python
len(X_train), len(X_test)
```

```python
import sklearn.linear_model as lm

# 학습할 준비, 바로 사용 시 .fit() 사용
reg = lm.LinearRegression().fit(X_train, y_train)
y_pred = reg.predict(X_test)
```

```python
import sklearn.metrics as mt

print("MSE =",mt.mean_squared_error(y_test, y_pred))

# 다중회귀에서는 R2 그대로보단 수정된 계수값을 쓰는 것이 좋기때문에
# R2를 구한 뒤 변수에 저장
r2 = mt.r2_score(y_test, y_pred)
print("R2 =",r2)

# n = 학습 데이터 전체 개수
# k = 독립변수 개수
n = len(X_train)
k = len(diab.feature_names)

adj_r2 = 1 - ((1-r2) * (n-1) / (n-k-1))
print("수정 R2 =", adj_r2)
```
